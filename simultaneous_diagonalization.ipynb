For n x n diagonalizable matrices, $\mathbf B$ and $\mathbf C$, then they commute *iff* they are simultaneously diagonalizable.  

**proof:**  
leg one  
$\text{simultaneous diagonalizability}\longrightarrow \mathbf{BC} = \mathbf{CB}$    
$\mathbf {BC} = \mathbf{SDS}^{-1}\mathbf {S\Lambda S}^{-1} = \mathbf{SD\Lambda S}^{-1} = \mathbf{S\Lambda DS}^{-1} = \mathbf {S\Lambda S}^{-1}\mathbf{SDS}^{-1}=\mathbf {CB}$  
because diagonal matrices commute 

leg two    
$\mathbf{BC} = \mathbf{CB} \longrightarrow \text{simultaneous diagonalizability}$    
idea: use a similarity transform to in effect assume WLOG that $\mathbf C$ is diagonal  

$\mathbf S^{-1}\big(\mathbf {BC}\big)\mathbf S = \mathbf S^{-1}\mathbf B\big(\mathbf{S\Lambda S}^{-1}\big)\mathbf S = \big(\mathbf S^{-1}\mathbf B\mathbf S\big) \mathbf\Lambda \mathbf S^{-1}\mathbf S = \big(\mathbf S^{-1}\mathbf B\mathbf S\big) \mathbf\Lambda = \mathbf X \mathbf \Lambda$  

where we have selected $\mathbf S$ so that $\Lambda$ is a diagonal matrix written such that if there are repeated eigenvalues, they are 'next to' each other, and in particular if there are any repeated roots, they are come before distinct ones. I.e. we have  


$\mathbf \Lambda = \begin{bmatrix}
\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 &  \mathbf 0 &  \cdots &  \mathbf 0 \\ 
\mathbf 0& \lambda_2 \mathbf I_{r^{(2)}} &  \mathbf 0 &  \cdots &   \mathbf 0 \\ 
 \mathbf 0&   \mathbf 0&  \lambda_3 \mathbf I_{r^{(3)}} &  \cdots  &  \mathbf 0 \\ 
 \mathbf 0&   \mathbf 0&   \mathbf 0&  \ddots &  \mathbf 0 \\ 
 \mathbf 0&   \mathbf 0 &  \mathbf 0&   \cdots&  \lambda_k \mathbf I_{r^{(k)}}
\end{bmatrix}$  

the nice case is when all eigenvalues are distinct, but using this blocked structure allows us to deal with repeated roots in a clean recursive manner.  So for Avoidance of doubt we have  

$\mathbf \Lambda = \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}$  

where if there are no repeated roots $\mathbf I_{r^{(1)}} =1$, but otherwise is a proper identity matrix (at least 2 x 2)  

so   
$\mathbf X = \big(\mathbf S^{-1}\mathbf B\mathbf S\big)$   

and

$\mathbf S^{-1}\big(\mathbf {CB}\big)\mathbf S = \mathbf S^{-1}\big(\mathbf{S\Lambda S}^{-1}\mathbf B\big)\mathbf S= \mathbf{\Lambda} \big(\mathbf S^{-1}\mathbf B\mathbf S\big) =\mathbf \Lambda \mathbf X$  

if they commute, then  
$\mathbf X \mathbf \Lambda=\mathbf S^{-1}\big(\mathbf {BC}\big)\mathbf S= \mathbf S^{-1}\big(\mathbf {CB}\big)\mathbf S=\mathbf \Lambda \mathbf X$  

but this is an eigenvector equation for diagonal matrix $\mathbf \Lambda$.  

The case we want to emphasize is that when all eigenvalues are distinct $\mathbf \Lambda$ has eigenspace dimension of one in each case, so the associated eigenvectors must be $\mathbf e_k \longrightarrow \mathbf X \text{ is diagonal}$  

for the nuisance case of repeated roots, we proceeds by blocks.  

Looking at our blocked structure we see  
$\begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\mathbf X  = \mathbf X \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}$  

implies   
$\begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix}  = \begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix} \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}$  

i.e. $\mathbf X$ itself must be block diagonal.  (This is dealt with indirectly in the block diagonalization problem  4.6.12 in Artin Chapter 4 notes.)    

i.e. if we consider the kernel of   
$\lambda_1\mathbf I_n - 
\begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}$

we see that its eigenspace consists *only* of the first $r^{(1)}$ standard basis vectors. Now by assumption $\mathbf X$ is diagonalizable (because $\mathbf B$ is), and from Artin chapter 4 we know that a block diagonal matrix is diagonalizable *iff* its blocks are, so we effect a similarity transform to see  

$\big(\mathbf V^{(1)}\big)^{-1}\begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix}\big(\mathbf V^{(1)}\big)$  
$=\begin{bmatrix} \mathbf V^{-1} &  \mathbf 0 \\ \mathbf 0& \mathbf I^{(1)}\\
\end{bmatrix} 
\begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf V &  \mathbf 0 \\ \mathbf 0& \mathbf I^{(1)}\\
\end{bmatrix}$  
$= \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf V^{-1}\mathbf X^{(\lambda_1)}\mathbf V &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix}$  
$= \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf D_1 &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix}$  
$= \begin{bmatrix} \mathbf D_1 &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix} \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}$  
$= \begin{bmatrix} \mathbf V^{-1} &  \mathbf 0 \\ \mathbf 0& \mathbf I^{(1)}\\
\end{bmatrix} \begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix} \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\begin{bmatrix} \mathbf V &  \mathbf 0 \\ \mathbf 0& \mathbf I^{(1)}\\
\end{bmatrix}$  
$=\big(\mathbf V^{(1)}\big)^{-1} \begin{bmatrix} \mathbf X^{(\lambda_1)} &  \mathbf 0 \\ \mathbf 0& \mathbf X^{(1)}\\
\end{bmatrix} \begin{bmatrix}\lambda_1 \mathbf I_{r^{(1)}} &  \mathbf 0 \\ \mathbf 0& \mathbf \Lambda^{(1)}\\
\end{bmatrix}\big(\mathbf V^{(1)}\big)$  



and if there are other repeated roots that concern us, we may recurse on  
$\mathbf \Lambda^{(1)}$  

(noting that for any matrix there are finitely many repeated roots and each time we recurse our problem becomes smaller by some integer amount)  

After recursing (and effecting incremental similarity transforms) as needed we have  
$\mathbf \Lambda \mathbf D = \mathbf D\mathbf \Lambda $  

i.e. we have  
$\mathbf \Lambda \mathbf D$  
$=\mathbf P^{-1}\Big(\mathbf B \mathbf C\Big)\mathbf P$  
$=\big(\mathbf V^{(l)}\big)^{-1}...\big(\mathbf V^{(2)}\big)^{-1}\big(\mathbf V^{(1)}\big)^{-1}\mathbf S^{-1}\Big(\mathbf B \mathbf C\Big)\mathbf S\big(\mathbf V^{(1)}\big)\big(\mathbf V^{(2)}\big)...\big(\mathbf V^{(l)}\big)  $  
$= \big(\mathbf V^{(l)}\big)^{-1}...\big(\mathbf V^{(2)}\big)^{-1}\big(\mathbf V^{(1)}\big)^{-1}\mathbf S^{-1}\Big(\mathbf C \mathbf B\Big)\mathbf S\big(\mathbf V^{(1)}\big)\big(\mathbf V^{(2)}\big)...\big(\mathbf V^{(l)}\big)$  
$=\mathbf P^{-1}\Big(\mathbf C \mathbf B\Big)\mathbf P$  
$=\mathbf D\mathbf \Lambda $  

and e.g. to make the point explicit we may re-write the last 2 equalities as  

$\mathbf P^{-1}\Big(\mathbf C \mathbf B\Big)\mathbf P= \mathbf P^{-1}\mathbf C \mathbf P \mathbf P^{-1}\mathbf B\mathbf P = \big(\mathbf P^{-1}\mathbf C \mathbf P \big)\big(\mathbf P^{-1}\mathbf B\mathbf P\big) =\big(\mathbf D\mathbf \Lambda\big) $  



The overall point is that through applying up to $n//2 + 1$ similarity transforms, we have simultaneously diagonalized $\mathbf {BC}$.  And since the composition of similarity transforms is a similarity transform we can run revisit out argument backwards and observe that $\mathbf P$ (and its inverse) maybe used to diagonalize both $\mathbf B$ and $\mathbf C$.  That is, for for diagonalizable $\mathbf B$ and $\mathbf C$  
$\mathbf{BC} = \mathbf{CB} \longrightarrow \text{simultaneous diagonalizability}$
