**claim** 

$\det(\mathbf A + \mathbf B) \geq \det(\mathbf A) + \det(\mathbf B)$

for Hermitian, positive definite $\mathbf A$ and positive semi-definite $\mathbf B$.  Both $\mathbf A$ and $\mathbf B$ are **n** x **n** matrices.  The writeup actively makes use of the fact that the determinant of a matrix is equal to the product of its eigenvalues.

Note: the original problems suggests that both matrices are positive semi-definite. To cover the special case where both $\mathbf A$ and $\mathbf B$ are singular, I will simply point out that the Right Hand Side (RHS) of the above inequality is zero.  The matrix given by $\big(\mathbf A + \mathbf B\big)$ would then be either full rank or not.  If not full rank, then it too has a determinant of zero, and hence the LHS = RHS.  If $\big(\mathbf A + \mathbf B\big)$ is full rank, then we know that it has a determinant $\gt 0$.  Why?  Because the addition of two Hermitian matrices results in a Hermitian matrix.  And in the case of both $\mathbf A$ and $\mathbf B$ being Hermitiation positive semi-definite, we know this means $\mathbf x^H \mathbf A \mathbf x \geq 0$  and $\mathbf x^H \mathbf B \mathbf x \geq 0$ for any $\mathbf x$, hence we can say that $\mathbf x^H \big(\mathbf A + \mathbf B) \mathbf x \geq 0$, which means that all eigenvalues of $\big(\mathbf A + \mathbf B\big)$ are themselve $\geq 0$.  The product of those eigenvalues must be $\geq 0$, and if the matrix $\big(\mathbf A + \mathbf B\big)$ is full rank then we know  $\det \big(\mathbf A + \mathbf B\big) \neq 0$, hence the product of those eigenvalues $\gt 0$, and the LHS is greater than the right hand side.  

The rest of this writeup assumes that at least one of the two Hermitian matrices is positive definite.  For simplicity, $\mathbf A$ will always be the **n** x **n** Hermitian matrix that is known to be positive definite, while $\mathbf B$ will be the the **n** x **n** Hermitian matrix that is said to be positive semi definite.


**proof**

$\mathbf A = \mathbf P \mathbf D \mathbf P^H = \mathbf P \mathbf D \mathbf P^{-1} = \mathbf A^H$  
$ \mathbf D_{k,k} =  \lambda_k \gt 0$

$\mathbf A^{\frac{1}{2}} = \mathbf P \mathbf D^{\frac{1}{2}} \mathbf P^H = \big(\mathbf A^{\frac{1}{2}}\big)^H$  
$ \mathbf D_{k,k}^{\frac{1}{2}}  =  \lambda_k^{\frac{1}{2}}  \gt 0$

Similarly for $\mathbf B$ we can say

$\mathbf B = \mathbf V \mathbf \Gamma \mathbf V^H = \mathbf V \mathbf \Gamma \mathbf V^{-1} = \mathbf B^H$  
$ \mathbf \Gamma_{k,k} = \gamma_k \geq 0$

$\mathbf B^{\frac{1}{2}} = \mathbf V \mathbf \Gamma^{\frac{1}{2}} \mathbf V^H = \big(\mathbf B^{\frac{1}{2}}\big)^H$  
$ \mathbf \Gamma_{k,k}^{\frac{1}{2}}  = \gamma_k^{\frac{1}{2}} \geq 0$

The above are all true because $\mathbf A$ and $\mathbf B$ are Hermitian positive (semi)definite, and hence we can nicely take the positive square root of each matrix.

$\det \big( \mathbf A^{\frac{-1}{2}} \big(\mathbf A + \mathbf B \big) \mathbf A^{\frac{-1}{2}} \big) \geq \det\big(\mathbf A^{\frac{-1}{2}} \mathbf A \mathbf A^{\frac{-1}{2}}  \big) + \det \big(\mathbf A^{\frac{-1}{2}}\mathbf B\mathbf A^{\frac{-1}{2}}\big)$

which can be equivalently written as 

$\det \big( \mathbf A^{-1} \big) \det\big(\mathbf A + \mathbf B \big) \geq \det\big(\mathbf A^{-1}\big) \det \big(\mathbf A\big) + \det \big(\mathbf A^{-1}\big) \det\big(\mathbf B \big)$

via properties of determinants

If the above inequality is true for all Hermitian positive definite $\mathbf A$ and all Hermitian positive semi-definite $\mathbf B$, then this means that I can take the above inequality and multipy everything by a positive scalar equal to $\\det \big( \mathbf A \big)$, proving that $\det\big(\mathbf A + \mathbf B\big) \geq \det(\mathbf A) + \det(\mathbf B)$. 

Since multiplying by a positive scalar does not change the inequality, proving 
$\\det \big( \mathbf A^{\frac{-1}{2}} \big(\mathbf A + \mathbf B \big) \mathbf A^{\frac{-1}{2}} \big) \geq \det\big(\mathbf A^{\frac{-1}{2}} \mathbf A \mathbf A^{\frac{-1}{2}}  \big) + \det \big(\mathbf A^{\frac{-1}{2}}\mathbf B\mathbf A^{\frac{-1}{2}}\big)$

$\det\big(\mathbf A + \mathbf B\big) \geq \det(\mathbf A) + \det(\mathbf B)$

So, we proceed onward.

Notice that $\big(\mathbf I + \mathbf \Sigma\big)$ is a diagonal matrix with $1 + \sigma_k$ in its kth diagonal entry. Put differently, $1 + \sigma_k$ is the kth eigenvalue of the matrix $\big(\mathbf I +  \mathbf\Sigma\big)$.

Thus  
$\det\big( \mathbf I + \mathbf \Sigma \big) \geq 1 + \det \big(\mathbf A^{-1}\big) \det \big(\mathbf B \big)$

This can be rewritten as 

$\prod_{k=1}^{n} \big(\sigma_k + 1\big) \geq \det \big(\mathbf A^{-1}\big) \det \big(\mathbf B \big) + 1$

$ \Big(\prod_{k=1}^{n} \sigma_k \Big) +  \Big(\prod_{k=1}^{n} 1 \Big) + \text{Sum Of Cross Terms} \geq \det \big(\mathbf A^{-1}\big) \det \big(\mathbf B \big) + 1$


$\Big( \det \big(\mathbf A^{-1}\big) \det \big(\mathbf B \big) \Big) + \Big(1\Big) + \text{Sum Of Cross Terms} \geq \Big(\det \big(\mathbf A^{-1}\big) \det \big(\mathbf B \big)\Big) + \Big(1\Big)$

$\text{Sum Of Cross Terms} \geq 0$

The fact that $\text{Sum Of Cross Terms} \geq 0$ is given by the fact that $1 \geq 0$, and $\sigma_k \geq 0$, hence every single cross terms is given by a product of scalars that are all $\geq 0$, and such a product is $\geq 0$, and the sum of finitely many numbers $\geq 0$ is itself $\geq 0$.  Put differently, because all cross terms are $\geq 0$, $\text{Sum Of Cross Terms} \geq 0$.

**claim:** 
$\det\big(\mathbf A + \mathbf B\big)^\frac{1}{n} \geq \det\big(\mathbf A\big)^\frac{1}{n} + \det\big(\mathbf B\big)^\frac{1}{n}$  

note that as in the above, we have

where $\mathbf C$ is still Hermitian and a quick quadratic form argument (or Sylvester's Law of Inertia) confirms that it is positive (semi) definite iff $\mathbf B$ is.


To prove the claim, we basically re-run the above argument and combine it with superadditivity of the geometric mean.  That is, on the LHS we have :  

**proof:** 

$LHS= \det\big(\mathbf A + \mathbf B\big)^\frac{1}{n} = \det\Big(\mathbf A^\frac{1}{2}\big(\mathbf I + \mathbf C\big)\mathbf A^\frac{1}{2}\Big)^\frac{1}{n} = \Big(\det\big(\mathbf A^\frac{1}{2}\big)\det\big(\mathbf I + \mathbf C\big)\det\big(\mathbf A^\frac{1}{2}\big)\Big)^\frac{1}{n}  = \det\big( \mathbf A^\frac{1}{2}\big)^\frac{1}{n} \det\big(\mathbf I + \mathbf C\big)^\frac{1}{n}\det\big( \mathbf A^\frac{1}{2}\big)^\frac{1}{n} $


and on the RHS we have 

$RHS = \det\big( \mathbf A^\frac{1}{2}\big)^\frac{1}{n}\Big(\det\big(\mathbf I\big)^\frac{1}{n} + \det\big(\mathbf C\big)^\frac{1}{n}\Big)\det\big( \mathbf A^\frac{1}{2}\big)^\frac{1}{n} = \Big(\det\big( \mathbf A^\frac{1}{2}\big)\det\big(\mathbf I\big)\det\big( \mathbf A^\frac{1}{2}\big)\Big)^\frac{1}{n} + \Big(\det\big( \mathbf A^\frac{1}{2}\big)\det\big(\mathbf C\big)\det\big( \mathbf A^\frac{1}{2}\big)\Big)^\frac{1}{n}$  
$= \det\big(\mathbf A\big)^\frac{1}{n} + \det\big(\mathbf B\big)^\frac{1}{n} $   


recalling that $\det\big( \mathbf A^\frac{1}{2}\big)^\frac{1}{n} \gt 0$ 

using the eigenvalues of $\mathbf C$, the claim reduces to:  

$\det\big(\mathbf I + \mathbf C\big)^\frac{1}{n} = \big(\prod_{j=1}^n (1+ \lambda_j)\big)^\frac{1}{n} = \text{GM}\big( \mathbf 1 + \mathbf \lambda \big) \geq \text{GM}\big(\mathbf 1\big)  + \text{GM}\big(\mathbf \lambda\big) =  \big(\prod_{j=1}^n 1\big)^\frac{1}{n}+\big(\prod_{j=1}^n \lambda_j\big)^\frac{1}{n} = \det\big(\mathbf I\big)^\frac{1}{n} + \det\big(\mathbf C\big)^\frac{1}{n}$ 

which holds due to the super-additivity of the geometric mean
