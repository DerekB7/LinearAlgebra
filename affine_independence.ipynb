This is the definition and 3 different ways to characterize affine (in)dependence of vectors.  This is used in particular on page 8 of Matousek's *Using the Borsuk Ulam Theorem* as in one or more of his *33 Miniatures* (e.g. miniature 8 of Packing Complete Bipartite Graphs)     


**definition**  
$\mathbf v_0, \mathbf v_1, ... , \mathbf v_k$ interpretted as 'points' $ \in \mathbb R^d$  are *affinely dependent*  if there are real numbers $\alpha_0, \alpha_1, ...,\alpha_k$, not identically 0 but $\sum_{i=0}^k \alpha_i =0$ and $\sum_{i=0}^k \alpha_i\mathbf v_i =\mathbf 0$. Otherwise $\mathbf v_0, \mathbf v_1, ... , \mathbf v_k$ are *affinely independent*.  

**remark**  
This looks a lot like linear independence except for the additional constraint that the $\alpha_i$'s sum to zero.  The right geometric way to interpret this is is that $\mathbf v_0$ is our 'new origin' and hence this is an affine translation of our vector space -- thus we call the result affine independence, not linear independence.  

**equivalent definition**  
$\mathbf v_1 - \mathbf v_0, ... , \mathbf v_k-\mathbf v_0$ are linearly dependent.  
**proof:**  
the above set are linearly dependent if there exists 

$\sum_{i=1}^k \alpha_i \big(\mathbf v_i - \mathbf v_0\big) = \mathbf 0$  
where at least one $\alpha_i \neq 0$.  But this means    

$\big(-\sum_{i=1}^k \alpha_i\big)\cdot \mathbf v_0  + \sum_{i=1}^k \alpha_i \mathbf v_i = \mathbf 0  $   
$\alpha_0 \cdot \mathbf v_0  + \sum_{i=1}^k \alpha_i \mathbf v_i = \mathbf 0  $   
$\sum_{i=0}^k \alpha_i \mathbf v_i = \mathbf 0  $   
where we select  
$\alpha_0 := \big(-\sum_{i=1}^k \alpha_i\big)$  

(i) Thus any time we satisfy this secondary definition, we satsify the original definition as we have a nontrival linear comination summing to zero and 
$\sum_{i=0}^k \alpha_i  = \alpha_0  + \sum_{i=1}^k \alpha_i = \big(-\sum_{i=1}^k \alpha_i\big)+ \sum_{i=1}^k \alpha_i  = 0$  

(ii) For any case deemed affinely dependent under our original definition, we show that it too is deemed so under the second definition,

The constraint 
$\sum_{i=0}^k \alpha_i = 0$  

is equivalent to 
$\alpha_0 = -\sum_{i=0}^k \alpha_i $  

*note:* this implies that aleast one $\alpha_i$ for $i \in \{1, 2, ... \}$ is nonzero for affine dependence  

supposing not all $\alpha_i = 0$  but 

$\sum_{i=0}^k \alpha_i\mathbf v_i =\mathbf 0$  
and the argument runs as before, but with the steps reversed   

$\alpha_0 \cdot \mathbf v_0  + \sum_{i=1}^k \alpha_i \mathbf v_i = \mathbf 0  $   
$\big(-\sum_{i=1}^k \alpha_i\big)\cdot \mathbf v_0  + \sum_{i=1}^k \alpha_i \mathbf v_i = \mathbf 0  $   
$\sum_{i=1}^k \alpha_i \big(\mathbf v_i - \mathbf v_0\big) = \mathbf 0$  


**further equivalent definition**  
The vectors   
$\begin{bmatrix}1\\ \mathbf{v}_0 \end{bmatrix}$,  $\begin{bmatrix}1 \\ \mathbf{v}_1 \end{bmatrix}$, ... , $\begin{bmatrix}1\\ \mathbf{v}_k \end{bmatrix}$  
are linearly dependent  

is equivalent to our original definition (and by transitivity, our 2nd definition as well)  

**proof**

we can encapsulate the agumented vectors in the below augmented matrix  

$\mathbf M := \begin{bmatrix}  \mathbf  1^T \\ \mathbf{V}\end{bmatrix}$  

with  
$\mathbf V := 
\bigg[\begin{array}{c|c|c|c|c}
\mathbf v_0 & \mathbf v_1 &\cdots & \mathbf v_{k-1} & \mathbf v_{k}
\end{array}\bigg]$   


Then $\mathbf M$ has linearly dependent columns *iff* there is some   

$\mathbf a = \begin{bmatrix}
\alpha_0 \\
\alpha_1 \\
\vdots\\ 
\alpha_{k-1} \\
\alpha_{k} 
\end{bmatrix} \neq \mathbf 0$  
such that  

$\mathbf{Ma} = \mathbf 0$  

but this implies 
$\mathbf {Va} = \mathbf 0$  
and

$\mathbf 1^T \mathbf a = 0 = \sum_{i=0}^k \alpha_i $  
which is our original definition  

hence all 3 definitions are equivalent
