assume $n\geq 2$, as there is nothing to do in the $n=1$ case  

let $f\in \mathbb F\big[\mathbf x\big]$ be $\frac{1}{2}$  symmetric (semi-symmetric) polynomial, i.e. invariant to even permutations for $\mathbf x$ where $\mathbf x$ has $x_i$ for $i\in\big\{1,2,..., n\big\}$.  

note: a skew-symmetric function is a refinement of this -- $g$ is a skew symmetric function iff 
$g\big(\sigma(\mathbf x)\big)=\text{sign}\big(\sigma\big)\cdot g\big(\mathbf x\big)$  
for every permutation in $S_n$-- or using matrix notation  
$g\big(P\mathbf x\big)=\det\big(P\big)\cdot g\big(\mathbf x\big)$  
for every $n\times n$ permutation matrix.  

(a) the square-root of the discriminant is skew symmetric because  
$\delta\big(\sigma(\mathbf x)\big) = \det\big(PW\big)=\det\big(P\big)\det\big(W\big)= \text{sign}\big(\sigma\big)\cdot \delta\big(\mathbf x\big)$  
where $W$ is the Vandermonde matrix associated with $\delta$ (ref ex 14.3.11)  

(b)  
*I assume* $\text{ char}\big(\mathbb F\big) =0$, though the problem was not explicit about this, and my understanding is the result holds over any field with characteristic $\neq 2$, and problems like this fall under the umbrella of "invariance theory".  

Now apply the symmetrization map (note since $n\geq 2$, by Lagrange's theorem, the even permutations subgroup of $S_n$ is exactly half the cardinality of $S_n$)   

$f_\text{sym}\big(\mathbf x\big) $  
$= \frac{1}{n!}\sum_{\sigma\in S} f\big(\sigma(\mathbf x)\big) $  
$= \frac{1}{n!}\Big(\sum_{\sigma_\text{even}\in S} f\big(\sigma_\text{even}(\mathbf x)\big)\Big)+\frac{1}{n!}\Big(\sum_{\sigma_\text{odd}\in S} f\big(\sigma_\text{odd}(\mathbf x)\big)\Big)$  
$= \frac{1}{n!}\Big(\sum_{\sigma_\text{even}\in S} f\big(\mathbf x\big)\Big)+\frac{1}{n!}\Big(\sum_{\sigma_\text{odd}\in S} f\big(\sigma_\text{odd}(\mathbf x)\big)\Big) $  
$= \frac{1}{n!}\cdot \frac{n!}{2}\cdot f\big(\mathbf x\big)+\frac{1}{n!}\Big(\sum_{\sigma_\text{odd}\in S} f\big(\sigma_\text{odd}(\mathbf x)\big)\Big) $  
$= \frac{1}{2} f\big(\mathbf x\big)+\Big(\frac{1}{n!}\sum_{\sigma_\text{odd}\in S} f\big(\sigma_\text{odd}(\mathbf x)\big)\Big) $  
$= \frac{1}{2} f\big(\mathbf x\big)+ p_\text{odd}\big(\mathbf x\big) $  
re-arranging terms  
$ f\big(\mathbf x\big) = 2\cdot f_\text{sym}\big(\mathbf x\big)-p_\text{odd}\big(\mathbf x\big)=f_\text{sym}\big(\mathbf x\big) + \underbrace{f_\text{sym}\big(\mathbf x\big)-p_\text{odd}\big(\mathbf x\big)} =f_\text{sym}\big(\mathbf x\big) + h\big(\mathbf x\big) $  

now we estimate the difference for $h\big(\mathbf x\big)$  
$h\big(\mathbf x\big)=f\big(\mathbf x\big)-f_\text{sym}\big(\mathbf x\big)$   

observe that for any $\sigma_\text{even}$, we have  
$h\big(\sigma_\text{even}(\mathbf x)\big)=f\big(\sigma_\text{even}(\mathbf x)\big)-f_\text{sym}\big(\sigma_\text{even}(\mathbf x)\big)=f\big(\mathbf x\big)-f_\text{sym}\big(\mathbf x\big)$   
thus $h$ **is a semi-symmetric function**    

- - - -  
The idea is to mimic the technique in Lemma 16.1.11 on page 480 of the *2nd Edition* and adapt it to the problem at hand.  

now using a well chosen homomorphism $\phi_{k,i}$ that sends $x_k\to x_i$ for $k\in \big\{2,3,...n\big\}$ for $i\in\big\{1,2,...,k-1\big\}$, and acts as the identity on everything else. In terms of interpretation, e.g. for sending $x_n\to x_i$ we may consider $\mathbf x':= 
\begin{bmatrix}
x_1\\  
x_2\\
\vdots \\
x_{n-1}\\
\end{bmatrix}$  and  
$R:= \mathbb F\big[\mathbf x'\big]$, so $\mathbb F\big[\mathbf x\big]\cong R[x_n]$ and this substitution homomorphism sends $R[x_n] \longrightarrow R $, and acts on the identity on $R$.  
- - - -  

notice that  
$\phi_{k,i}\Big(f\big(\mathbf x\big)\Big)= \phi_{k,i}\Big(f\big(\sigma_\text{even}(\mathbf x)\big)\Big)= \phi_{k,i}\Big(f\big(\sigma_\text{odd}(\mathbf x)\big)\Big)$  
where the first equality follows because $f$ is semi-symmetric, so $f\big(\mathbf x\big) =f\big(\sigma_\text{even}(\mathbf x)\big)$ and in turn the image under a homomorphism is the same, and the second equality follows because $\phi_{k,i}\Big(f\big(\sigma_\text{odd}(\mathbf x)\big)\Big)=\phi_{k,i}\Big(f\big(\sigma_\text{some even}(\mathbf x)\big)\Big)$, that is any odd permutation is equivalent under the homomorphism to some even permutation --i.e. that exact same odd permutation composed with a basic transposition (pairwise swap) of elements $k$ and $i$.    

Thus, $\phi_{k,i}\Big(f\big(\mathbf x\big)\Big)=\phi_{k,i}\Big(f_\text{sym}\big(\mathbf x\big)\Big)$  
$\implies \phi_{k,i}\Big(h\big(\mathbf x\big)\Big) = 0$,  
so $h$ is a multiple of $\big(x_k-x_i\big)$  

after re-running this argument for all possible pairs of $k$ and $i$ we get  

$h\big(\mathbf x\big)=\delta\big(\mathbf x\big)\cdot g_1\big(\mathbf x\big)$  
where $\delta$ is the square root of the discriminant / Vandermonde determinant for $\mathbf x$.  Note the sign of $+1$ or $-1$ doesn't matter here-- if it were $-1$ we could just re-label $-1 \cdot g_1$ to be $g_1$-- this does not affect the end result. $g_1$ is some as of yet unknown polynomial.

*to finish*  
with $n$ fixed, we proceed by induction on degree $d$ (with convention that zero polynomial has degree $-\infty$).  

$\text{degree}\Big(f\big(\mathbf x\big)\Big) = d =\text{degree}\Big(f_\text{sym}\big(\mathbf x\big)\Big)$   
$\implies \text{degree}\Big(h\big(\mathbf x\big)\Big)=\text{degree}\Big(f\big(\mathbf x\big)-f_\text{sym}\big(\mathbf x\big)\Big)\leq d$   

First note that $\text{degree}\Big(\delta\Big)\geq n-1$ and since the ground ring $\mathbb F$ is a field and hence an integral domain, we know $d\geq n-1\implies g_1\in \mathbb F$ which is trivially a symmetric polynomial. 

Thus we have a base case that for all $d\leq n-1$, it is true that 
$ f\big(\mathbf x\big)= q\big(\mathbf x\big)+ \delta\cdot m\big(\mathbf x\big)$  
where $q,m$ are symmetric polynomials.  

Now begin the strong induction:   
Suppose the desired factorization is true for degrees up to $d-1$ for any $d\geq n$.  Then by the above argument, we have  
$h\big(\mathbf x\big)=\delta\big(\mathbf x\big)\cdot g_1\big(\mathbf x\big)$   
and $\text{degree}\big(h\big)\leq d \implies \text{degree}\big(g_1\big)\leq d-(n-1)$  

since $h$ is a semi-symmetric polynomial, and $\delta$, being skew symmetric, is semi-symmetric as well, which implies $g_1$ is semi-symmetric.  Calling on our induction hypothesis for all lower degree semi-symmetric polynomials in $n$ variables, we have  
$g_1\big(\mathbf x\big) = m_1\big(\mathbf x\big) + \delta\big(\mathbf x\big)\cdot m_2\big(\mathbf x\big)$  
where $m_1$ and $m_2$ are symmetric functions.  To close,   


$ f\big(\mathbf x\big)$  
$=f_\text{sym}\big(\mathbf x\big) + h\big(\mathbf x\big) $  
$=  f_\text{sym}\big(\mathbf x\big)+ \delta\big(\mathbf x\big)\cdot g_1\big(\mathbf x\big)$  
$=  f_\text{sym}\big(\mathbf x\big)+ \delta\big(\mathbf x\big)\cdot\Big(m_1\big(\mathbf x\big) + \delta\big(\mathbf x\big)\cdot m_2\big(\mathbf x\big)\Big)$  
$=  \underbrace{f_\text{sym}\big(\mathbf x\big)+ \delta\big(\mathbf x\big)^2\cdot m_2\big(\mathbf x\big)} + \delta\big(\mathbf x\big)\cdot m_1\big(\mathbf x\big)  $   
$=  q\big(\mathbf x\big)+ \delta\cdot m_2\big(\mathbf x\big)  $   

where the discriminant $\delta\big(\mathbf x\big)^2$ is a symmetric function and a symmetric function times a symmetric function is a symmetric function, and the sum of symmetric functions is a symmetric function (that is from section 3 we recall the notion of the *ring of symmetric functions* hence there is closure under multiplication and addition), hence $q$ is a symmetric function and $f$ is written as the sum of a symmetric function plus $\delta$ times a symmetric function, as desired.  



**14.Misc.10**  
Let $\mathbb K$ be a finite extension of a field $\mathbb F$, and let $f(x)\in \mathbb K[x]$.  Prove that there is a non-zero $g(x)\in \mathbb K[x]$ such that $f(x)g(x)\in \mathbb F[x]$.  

**remark:**  
to simplify the argument and notation, I assume $\text{char}\big(\mathbb F[x]\big) = 0$.  This allows us to make use of the primitive element theorem (4.1).  After understanding the below argument, one should be able to adapt it to the case of arbitrary fields, though as stated on p. 537 (first page chp 14), we in general restrict our attention to fields of characteristic zero for this chapter.  

**note:**  
at the risk of some confusion, we contemplate 'scalars' in two different forms in the below argument.  One is in the sense of a scalar $\in \mathbb F$ our ground field, and the other is from our 'base ring' $R:=\mathbb F[x]$.  

**proof:**  
$\big[\mathbb K:\mathbb F\big]=d$.  Let $\lambda_1$ be the primitive element generating the extension, i.e.  
$\mathbb K = \mathbb F(\lambda_1)$. And let $h(x)\in\mathbb F[x]$ be the monic irreducible polynomial (of degree d) which generates this extension (i.e. that has $\lambda_1$ as a root).  

For convenience define $\mathbb K'$ as the splitting field of $h(x)$ so $\mathbb F\subseteq \mathbb K \subseteq \mathbb K'$.   

with $a_i \in \mathbb K[x]$   
$f(x)= a_n + a_{n-1}x+... +a_1x^{n-1} + a_0 x^n$  
and since $\lambda_1$ generates the extension, each $a_i$ can be written as a polynomial in $\lambda_1$ with scalars from $\mathbb F$.  That is   
$f(x)= a_n + a_{n-1}x+... +a_1x^{n-1} + a_0 x^n = p_n\big(\lambda_1\big)+p_{n-1}\big(\lambda_1\big)x+...+p_1\big(\lambda_1\big)x^{n-1}+p_0\big(\lambda_1\big)x^{n}$   
thus $f(x) \in \mathbb F\big[x,\lambda_1\big]\cong \mathbb F\big[x\big]\big[\lambda_1\big]= R[\lambda_1]$  
that is, we can interpret $f(x)$ as being a polynomial in $\lambda_1$ with "scalars" from our base ring $R$.  

now, temporarily working over the splitting field $\mathbb K'$:   

$h(x)$ has $d$ roots (they are necessarily distinct since we work over characteristic zero, but we don't need this), which we call $\big\{\lambda_2, \lambda_3,...,\lambda_d\big\}$  

for $k\in \big\{2,3,...,d\big\}$  
$g_k(x) :=  p_n\big(\lambda_k\big)+p_{n-1}\big(\lambda_k\big)x+...+p_1\big(\lambda_k\big)x^{n-1}+p_0\big(\lambda_k\big)x^{n}$  
where each $g_k$ maybe be interpreted as living in $R[\lambda_k]$

$g\big(x\big):= \prod_{k=2}^d g_k(x)$  

and finally  
$q\big(x\big):=f\big(x\big)\cdot g\big(x\big)$  
where we can interpret $f\big(x\big)\cdot g\big(x\big) \in R\big[\lambda_1, \lambda_2,....,\lambda_d\big]$  
and so $q$, there product, lives in $R\big[\lambda_1, \lambda_2,....,\lambda_d\big]$ as well.  

*(i)* we claim $q\in \mathbb F[x]$.  This holds because 
$q\in R\big[\lambda_1, \lambda_2,....,\lambda_d\big]$  
and $q$ is a *symmetric polynomial* of all $d$ of these $\lambda_k$, and per 3.4, elementary symmetric function theorem (p.548), $q$ may be written in a unique ways as a polynomial of the elementary symmetric functions of $\lambda_k$.  The elementary symmetric functions of $\lambda_k$ give (up to a sign) coefficients of the polynomial $h(x)\in \mathbb F\big[x\big]$, so they live in $\mathbb F\big[x\big]$, hence $q$ does as well.  

*(ii)*  since $q\in \mathbb F[x]\subseteq \mathbb K\big[x\big]$ and  $f \in \mathbb K\big[x\big]$ and   
$q\big(x\big)=f\big(x\big)\cdot g\big(x\big)$   
this means $g\big(x\big)\in \mathbb K\big[x\big]$ as well.  

Justification: working in $\mathbb K\big[x\big]$, we can do basic polynomial division (since $\mathbb K$ is a field), and after we factor $f(x)$ from $q(x)$ we get $g(x)\in \mathbb K[x]$ (since factoring is closed with respect to our ring  $\mathbb K[x]$  

alternative justification: using principal ideal, compute the same thing 2 different ways    
$p(x)\in \big(f(x)\big)$ -- i.e. the principal ideal in $\mathbb K\big[x\big]$ generated by $f(x)$. So $p(x)=f(x)\cdot z(x)$ for $z(x) \in \mathbb K[x]$.  And working over the splitting field $\mathbb K'[x]$, we know $p(x)=f(x)\cdot z(x)$ is still true, but we also know $p(x)=f(x)\cdot g(x)$.  So $f(x)\cdot \big(g(x)-z(x)\big)=0$ but $\mathbb K'[x]$ is in integral domain, so there are no zero divisors $\implies g(x) = z(x)\in \mathbb K[x]$.  


miscellaneous calculations, below

```python
import sympy as sp
y = sp.Symbol('y')

def sigma(x):
    num = x + sp.I
    denom = x - sp.I
    return num / denom

result = y

for _ in range(4):
    print(result," and ", _)
    result = sigma(result)
    result = sp.simplify(result)

def tau(x):
    num = sp.I*(x -1)
    denom = x+1
    return num / denom
    
print(sp.simplify(tau(sigma(y))))
print(sp.simplify(sigma(tau(y))))
# so this composition is an element of of the automorphism group... and it has order 2   

result = y
for _ in range(4):
    print(result," and ", _)
    result = tau(result)
    result = sp.simplify(result)

thing = tau(tau(tau(result))) - y
print(thing.simplify())
```
