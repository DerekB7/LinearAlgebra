# Key Lemma
Both of the following proofs use the fact that the trace of a scalar is that same scalar, and for any two vectors of dimension $r$, 

$trace\big(\tilde{ \mathbf x}^T \mathbf y\big) =\tilde{ \mathbf x}^T \mathbf y = x_1 y_1 + x_2 y_2 + ...  + x_{r-1} y_{r-1}  + x_r y_r = trace\big( \mathbf y \tilde{ \mathbf x}^T \big)$

which we may verify by inspection, recalling that the rank one matrix given by $ \mathbf y \tilde{ \mathbf x}^T $ is  

$\mathbf y \tilde{ \mathbf x}^T =  \begin{bmatrix}y_1 \tilde{ \mathbf x}^T \\
y_2 \tilde{ \mathbf x}^T \\ 
\vdots\\ 
y_{r-1} \tilde{ \mathbf x}^T \\ 
y_r \tilde{ \mathbf x}^T
\end{bmatrix}  = \begin{bmatrix} y_{1}x_{1} & y_{1}x_{2} & \cdots & y_{1}x_{r-1} & y_{1}x_{r} \\ 
y_{2}x_{1} & y_{2}x_{2} & \cdots & y_{2}x_{r-1}  & y_{2}x_{r} \\ 
\vdots & \vdots &  \ddots & \vdots & \vdots \\ 
y_{r-1}x_{1} & y_{r-1}x_{2} & \cdots & y_{r-1}x_{r-1} & y_{r-1}x_{r} \\ 
y_{r}x_{1} & y_{r}x_{2} & \cdots & y_{r}x_{r-1} & y_{r}x_{r}
\end{bmatrix} = \begin{bmatrix} y_{1}x_{1} & * & \cdots & * &* \\ 
* & y_{2}x_{2} & \cdots &*  &* \\ 
\vdots & \vdots &  \ddots & \vdots & \vdots \\ 
* & * & \cdots & y_{r-1}x_{r-1} & * \\ 
* & * & \cdots & * & y_{r}x_{r}
\end{bmatrix}$

where a $*$ as an entry indicates that we unconcerned with the value in that cell


# Proof 1

where $\mathbf A$ is m x n and $\mathbf B $ is n x m

$\mathbf A = 
\bigg[\begin{array}{c|c|c|c|c}
\mathbf a_1 & \mathbf a_2 &\cdots & \mathbf a_{n-1} & \mathbf a_{n}
\end{array}\bigg]
$   
$\mathbf B= 
\begin{bmatrix}
\tilde{ \mathbf b_1}^T \\
\tilde{ \mathbf b_2}^T \\ 
\vdots\\ 
\tilde{ \mathbf b}_{n-1}^T \\ 
\tilde{ \mathbf b_n}^T
\end{bmatrix}
$   

- - - -
now consider that  
$\mathbf{AB}  = \mathbf a_1 \tilde{ \mathbf b_1}^T + \mathbf a_2 \tilde{ \mathbf b_2}^T+ ... + \mathbf a_{n-1} \tilde{ \mathbf b}_{n-1}^T + \mathbf a_n \tilde{ \mathbf b_n}^T $


$\mathbf{BA} = \begin{bmatrix}
\tilde{ \mathbf b_1}^T \mathbf a_1 & * & \cdots & * &* \\ 
* & \tilde{ \mathbf b_2}^T \mathbf a_2 & \cdots &*  &* \\ 
\vdots & \vdots &  \ddots & \vdots & \vdots \\ 
* & * & \cdots & \tilde{ \mathbf b}_{n-1}^T \mathbf a_{n-1} & * \\ 
* & * & \cdots & * & \tilde{ \mathbf b_n}^T \mathbf a_n
\end{bmatrix}$

hence 

$trace\big(\mathbf{BA}\big) = \tilde{ \mathbf b_1}^T \mathbf a_1 + \tilde{ \mathbf b_2}^T \mathbf a_2 + ... + \tilde{ \mathbf b}_{n-1}^T \mathbf a_{n-1} + \tilde{ \mathbf b_n}^T \mathbf a_n $


$trace\big(\mathbf{BA}\big) = trace\big(\tilde{ \mathbf b_1}^T \mathbf a_1\big) + trace\big(\tilde{ \mathbf b_2}^T \mathbf a_2\big) + ... +  trace\big(\tilde{ \mathbf b}_{n-1}^T \mathbf a_{n-1}\big) + trace\big(\tilde{ \mathbf b_n}^T \mathbf a_n\big) $

$trace\big(\mathbf{BA}\big) = trace\big(\mathbf a_1 \tilde{ \mathbf b_1}^T\big) + trace\big(\mathbf a_2 \tilde{ \mathbf b_2}^T\big) + ... + trace\big(\mathbf a_{n-1} \tilde{ \mathbf b}_{n-1}^T\big) + trace\big(\mathbf a_n \tilde{ \mathbf b_n}^T\big)$

$trace\big(\mathbf{BA}\big) = trace\big( \mathbf a_1 \tilde{ \mathbf b_1}^T + \mathbf a_2 \tilde{ \mathbf b_2}^T + ... +  \mathbf a_{n-1} \tilde{ \mathbf b}_{n-1}^T + \mathbf a_n \tilde{ \mathbf b_n}^T\big)$

$trace\big(\mathbf{BA}\big) = trace\big( \mathbf{AB}\big)$



# Proof 2


*commentary: *
This was a suggested approach in Linear Algebra Done Wrong.  To a large extent this proof leverages linearity of the traces, and in effect -- indicator variables -- in a way reminiscent of linearity of expectations.  An alternartive view is that we are taking advantage of linearity and looking at a basis for matrices when we view matrices as a vector space

*setup: *  
now consider

$\mathbf X_{i,j}$ which is a matrix comprised entirely of zeros, except there is a $1$ in cell $(i,j)$

now observe that any matrix $\mathbf B$ that is n x m can be written as a linear combination of $\mathbf X$'s, i.e.:

$\mathbf B = \sum_{i=1}^n \sum_{j=1}^m \gamma_{i,j} \mathbf X_{i,j} $

where $\gamma_{i,j}$ is the appropriate scalar.  


specifically,

if we can prove, for some m x n matrix that we call $\mathbf A$, 

$trace\big(\mathbf{A}\mathbf X_{i,j}\big) = trace\big(\mathbf X_{i,j}\mathbf{A}\big) $

then we automatically get 

$trace\big(\mathbf{BA}\big) = trace\big(\mathbf{AB}\big) $

because

(extracted from page 23 of 'Linear Algebra Done Wrong' with minor adjustment)

>$T (X) = trace(AX)$,
>$T_1 (X) = trace(XA)$  
>To prove the theorem it is sufficient to show that $T = T_1$ ; the equality for
>X = B gives the theorem.  
>Since a linear transformation is completely defined by its values on a
>generating system, we need just to check the equality on some simple matrices, 
>for example on matrices $X_{i,j}$, which has all entries 0 except the entry
>1 in the intersection of jth column and [i]th row.

*alternative rationale*

we verify the above argument via direct calculation.  The result ultimately comes down to using linearity of the trace, and interchanging it with summations as needed (which is always legal in finite series, i.e. when there aren't special convergence issues to worry about).  

$trace\big(\mathbf{BA}\big) = trace\Big(\big(\sum_{i=1}^n \sum_{j=1}^m \gamma_{i,j} \mathbf X_{i,j}\big) \mathbf A   \Big) = trace\Big(\sum_{i=1}^n \sum_{j=1}^m \gamma_{i,j} \mathbf X_{i,j}\mathbf A   \Big) = \sum_{i=1}^n \sum_{j=1}^m  trace\Big(\gamma_{i,j}\mathbf X_{i,j} \mathbf A   \Big)$

$trace\big(\mathbf{BA}\big) = \sum_{i=1}^n \sum_{j=1}^m \gamma_{i,j} trace\Big(\mathbf X_{i,j} \mathbf A   \Big) = \sum_{i=1}^n \sum_{j=1}^m \gamma_{i,j} trace\Big( \mathbf A \mathbf X_{i,j}  \Big)= \sum_{i=1}^n \sum_{j=1}^m  trace\Big( \mathbf A \big(\gamma_{i,j}\mathbf X_{i,j}  \big)\Big)$

$ trace\big(\mathbf{BA}\big) = trace\Big(\sum_{i=1}^n \sum_{j=1}^m  \mathbf A \big(\gamma_{i,j}\mathbf X_{i,j}  \big)\Big) = trace\Big( \mathbf A \big( \sum_{i=1}^n \sum_{j=1}^m  \gamma_{i,j}\mathbf X_{i,j}  \big)\Big) =  trace\big(\mathbf{AB}\big)$

**main proof**


evaluate by calculation:  

$trace\Big(  \mathbf X_{i,j}\mathbf{A} \Big)$

First observe that we are only interested in the diagonal entries of $\big(\mathbf X_{i,j}\mathbf{A} \big)$ and that every diagonal entry not in row $i$ must be zero.  We now expand the multiplication of the ith row of $\mathbf X$ and the ith column of $\mathbf A$

$trace\Big(  \mathbf X_{i,j}\mathbf{A} \Big) = \sum_{k=1}^m (x_{i,k}) a_{k,i} = (0) a_{1,i} + ... + (0) a_{j-1,i} + (1) a_{j,i} +  (0) a_{j + 1,i} + ... +  (0) a_{m,i} =  a_{j,i} $

and now evaluate by calculation:  

$trace\Big( \mathbf{A}\mathbf X_{i,j} \Big)$

We are only interested in the diagonal entries of $\big( \mathbf{A}\mathbf X_{i,j} \big)$ and observe that every diagonal entry not in column $j$ must be zero.  We now expand the multiplication of the jth row of $\mathbf A$ and the jth column of $\mathbf X$

$trace\Big( \mathbf{A}\mathbf X_{i,j} \Big) =   \sum_{k=1}^n a_{j,k}(x_{k,j})= a_{j,1}(0) + ... + a_{j,i-1}(0) + a_{j,i}(1) +  a_{j,i + 1}(0) + ... +   a_{j,n}(0) =  a_{j,i} $

- - - -
**alternative proof**

recalling that $\mathbf X_{i,j}$ which is a matrix comprised entirely of zeros, except there is a $1$ in cell $(i,j)$, we can rewrite it using standard basis vectors, specifically where :



$\mathbf I_n = 
\bigg[\begin{array}{c|c|c|c}
\mathbf e_1 & \mathbf e_2 &\cdots & \mathbf e_{n}
\end{array}\bigg]
$   

$\mathbf I_m = 
\bigg[\begin{array}{c|c|c|c}
\mathbf u_1 & \mathbf u_2 &\cdots & \mathbf u_{m}
\end{array}\bigg]
$   


$\mathbf X_{i,j} = \mathbf e_i \mathbf u_j^T $


$trace\Big(\mathbf{A}\mathbf X_{i,j}\Big)  = trace\Big(\mathbf{A} \mathbf e_i \mathbf u_j^T \Big) = trace\Big(\big(\mathbf{A} \mathbf e_i\big) \mathbf u_j^T \Big)  = trace\Big( \mathbf u_j^T \big(\mathbf{A} \mathbf e_i\big)\Big)  = trace\Big( \mathbf u_j^T \mathbf{A} \mathbf e_i\Big) = a_{j,i} $

$ trace\Big(  \mathbf X_{i,j}\mathbf{A} \Big) = trace\Big(  \mathbf e_i \mathbf u_j^T \mathbf{A} \Big) = trace\Big(  \mathbf e_i \big(\mathbf u_j^T \mathbf{A}\big) \Big)  = trace\Big( \big(\mathbf u_j^T \mathbf{A}\big)  \mathbf e_i \Big)  = trace\Big( \mathbf u_j^T \mathbf{A} \mathbf e_i \Big) = a_{j,i}$
- - - -
In each case we conclude 

$trace\Big(\mathbf{A}\mathbf X_{i,j}\Big)  = trace\Big(  \mathbf X_{i,j}\mathbf{A} \Big)$ 

which completes the proof
